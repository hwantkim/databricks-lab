# 강의 - Databricks에서 AI 에이전트와 도구의 기초

## 개요

이 강의는 Databricks Platform 내에서 AI 에이전트와 도구의 기본 개념을 소개합니다. 현대의 AI 애플리케이션은 데이터와 상호작용하고 분석 작업을 수행하며 이용 가능한 정보에 기반한 결정을 내릴 수 있는 에이전트를 필요로 합니다. 이러한 기본 개념을 이해함으로써 거버넌스, 보안, 분석 역량을 결합한 견고하고 확장 가능한 AI 솔루션을 구축할 준비가 될 것입니다.

AI 에이전트는 단순히 사용자 프롬프트에 따라 정보를 제공하는 전통적인 AI 시스템에서 혁신적인 변화를 나타냅니다. 대신, 에이전트들은 이용 가능한 도구를 활용해 더 정확하고 정보에 기반한 의사결정을 내리고, 환경 내에서 자율적으로 행동하여 사용자 정의 목표를 달성합니다.

### 학습 목표

이 강의가 끝나면 다음을 할 수 있을 것입니다:

- AI 에이전트가 무엇인지 정의하고 그 핵심 구성 요소와 기능을 이해한다
- AI 에이전트 아키텍처에서 도구의 역할과 에이전트 기능을 확장하는 방법을 설명하다.
- 에이전트 도구 거버넌스 및 관리를 위해 Unity Catalog 사용의 이점을 확인하다.

## A. AI 에이전트 이해하기

### A1. AI 에이전트란 무엇인가요?

**AI 에이전트**는 환경을 인지하고 의사결정을 내리며 특정 목표를 달성하기 위한 행동을 취할 수 있는 지능형 소프트웨어 시스템입니다. 사용자의 지속적인 입력이 필요한 전통적인 AI 시스템과 달리, AI 에이전트는 다음과 같은 자율적 시스템입니다:

- 복잡한 문제와 상황에 대한 **추론**
- 목표 달성을 위한 일련의 행동 순서를 **계획**
- 새로운 정보에 따른 행동 **조정 및 적용**
- 외부 시스템 및 데이터 소스와 **상호작용**
- 향후 성능 향상을 위한 경험을 통한 **향상**

AI 에이전트가 주목받는 이유는 바로 **적응성** 때문입니다. 에이전트는 도구를 사용해 최신 데이터 세트를 동적으로 가져와 의사 결정과 프로세스에 반영하므로, 복잡하고 예측 불가능한 작업에 매우 적합합니다. 인간이 목표를 설정하면, AI 에이전트는 그 목표를 달성하기 위한 최선의 방법을 스스로 결정합니다.

데이터 분석 및 비즈니스 인텔리전스(BI) 분야에서 AI 에이전트는 사용자와 데이터 시스템 사이에서 지능형 중개자 역할을 하며, 자연어 질의를 이해하고 복잡한 분석 작업을 수행할 수 있습니다.

### A2. AI 에이전트의 진화

AI 에이전트는 등장 이후 비약적으로 발전해 왔습니다.

- **1960년대 - 규칙 기반 시스템**
    - 미리 정해진 논리 트리를 가진 기본 챗봇
    - 경직된 규칙 기반 프로그래밍
    - 단순하고 시나리오에 기반한 답변만 가능

- **1990년대 - 통계적 학습**
    - 정보를 처리하는 자율성이 향상된 시스템
    - 단순한 의사 결정 능력 보유
    - 소비자용 AI 기기의 기반 마련

- **2000년대 - 머신 러닝 통합**
    - 로봇 청소기, 디지털 비서(Siri, Alexa) 같은 소비자용 기기 등장
    - 통계적 머신러닝 모델 및 인공신경망 활용
    - 향상된 의사결정 및 분석 능력

- **2020년대 - 대규모 언어 모델**
    - 심층 강화 학습 및 트랜스포머 기반 대규모 언어 모델(LLM)의 획기적 발전
    - 멀티모달 인터페이스 및 고도화된 추론 능력
    - 복잡한 환경과의 동적 상호작용
    - 기능 확장을 위한 툴 호출(Tool-calling) 능력

### A3. AI 에이전트의 주요 원칙

AI 에이전트는 전통적인 소프트웨어와 구별되는 세 가지 기본 원칙에 따라 작동합니다:

#### **인식(Perception)**
에이전트가 자신이 작동하는 맥락을 이해하기 위한 첫 번째 단계입니다. 언어 모델의 경우 다음 사항들이 여기에 포함됩니다.
- 텍스트, 사진, 음성을 통한 사용자의 입력 및 질문
- 센서 또는 APIs에서 얻은 환경 데이터
- 과거 이력 맥락 및 대화 기록

#### **의사결정(Decision-Making)**
에이전트는 수집된 정보를 알고리즘을 통해 처리하고 사용자 목표에 맞춰 적절한 행동을 결정합니다:
- 요구사항 및 제약 조건 분석
- 필요한 단계와 도구 사용 결정
- 최적 실행 순서 계획

#### **행동(Action)**
마지막으로, 에이전트는 목표를 달성하기 위해 구체적인 조치를 취합니다:
- 데이터베이스 쿼리 및 API 호출 실행
- 데이터 처리 및 변환
- 보고서 및 권고사항 생성
- 실제 결과에 영향을 미치는 의사결정 내리기

### A4. AI 에이전트의 핵심 구성 요소
현대의 AI 에이전트는 일반적으로 여러 핵심 구성 요소로 구성되어 함께 작동합니다:

1. **대형 언어 모델(LLM) 브레인**
자연어를 처리하고 문맥을 이해하며, 어떤 행동을 취할지 결정하는 중앙 추론 엔진입니다.
1. **메모리 시스템**
대화 기록, 문맥, 학습된 정보를 저장하여 시간이 지나도 일관된 상호작용을 유지할 수 있도록 합니다.
1. **기획 모듈**
복잡한 요청을 더 작고 관리 가능한 작업으로 분해하고 최적의 행동 순서를 결정합니다.
1. **도구 인터페이스**
에이전트를 외부 시스템, 데이터베이스, API, 기능 등과 연결하여 텍스트 생성 이상의 능력을 확장해 줍니다.
1. **실행 엔진**
계획된 행동의 실제 실행을 관리하고 외부 도구 및 시스템에서 오는 응답을 처리합니다.

### A5. 복잡도에 따른 AI 에이전트 유형

AI 에이전트는 복잡성과 적용 분야에 따라 다양하게 분류됩니다. 이러한 유형을 이해하면 특정 사용 사례에 맞는 적절한 접근 방식을 선택하는 데 도움이 됩니다.
1. **단순 반사 에이전트(Simple Reflex Agents)**
    - 오직 현재의 조건만을 기반으로 의사를 결정함
    - 예시: 먼지를 감지했을 때만 청소하는 로봇 청소기
    - 과거의 이력이나 미래에 미칠 영향은 고려하지 않음
1. **모델 기반 반사 에이전트(Model-Based Reflex Agents)**
    - 현재 상태를 파악하고 환경 예측 모델(world model)을 활용하여 행동을 제어함
    - 예시: 시간, 날씨, 사용자 선호도를 바탕으로 온도를 조절하는 스마트 서모스탯(온도 조절기)
    - 단순 반사 에이전트보다 더 정교함
1. **목표 기반 에이전트(Goal-Based Agents)**
    - 원하는 목표를 달성하기 위해 구체적인 전략을 계획함
    - 행동 순서를 수립하고 진행 상황을 평가함
    - 예시: 교통 상황과 경로를 고려하는 구글 맵과 같은 네비게이션 시스템
1. **유틸리티 기반 에이전트(Utility-Based Agents)**
    - 최적의 효율성을 위해 목표를 달성하는 여러 방법을 평가함
    - 위험 대비 보상 모델과 최적화 기준을 고려함
    - 예시: 투자 전략을 조정하는 AI 트레이딩 봇
1. **학습 에이전트(Learning Agents)**
    - 과거의 행동으로부터 학습하고 미래의 상황에 적응함
    - 성과를 분석하고 효율성 개선을 도모함
    - 예시: 사용자 행동을 바탕으로 성능이 향상되는 추천 시스템

### A6. 모든 LLM이 도구를 사용할 수 있나요?
아니요, 모든 LLM이 도구 호출 기능을 가진 것은 아닙니다. Databricks에서는 LLM의 도구 사용이 Databricks Assistant 또는 외부 시스템, 데이터베이스, 또는 APIs와 상호작용할 수 있도록 하는 맞춤형 에이전트 프레임워크와 같은 특정 프레임워크와 통합을 통해 가능해집니다. 이 기능은 모든 LLM에 내재된 것은 아니며, 이는 안전하고 효과적인 도구 사용을 보장하기 위해 추가적인 엔지니어링, 오케스트레이션, 보안 통제가 필요합니다. 예를 들어, Databricks Assistant은 Databricks 환경 내에서 질문에 답하고 작업을 수행하는 도구를 사용하도록 설계되었지만, 이는 플랫폼의 기능일 뿐 모든 LLM의 보편적인 기능은 아닙니다. 

> 도구 호출이 가능한 기반 모델(Foundation Model) API의 전체 목록을 확인하시려면 [여기](https://docs.databricks.com/aws/en/machine-learning/model-serving/function-calling#supported-models)에서 자세한 내용을 읽어보시기 바랍니다.

## B. 에이전트 도구 이해하기

### B1. 에이전트 도구란 무엇인가요?

**에이전트 도구**는 AI 에이전트가 외부 시스템과 상호작용하고 특정 작업을 수행할 수 있도록 확장하는 특수한 기능이나 역량입니다. 도구를 AI 에이전트의 '손'으로 생각해보세요. LLM이 추론과 의사결정을 위한 '두뇌'를 제공하는 반면, 도구는 에이전트가 실제로 데이터를 조작하고, APIs을 호출하며, 계산을 수행하고, 현실 세계와 상호작용할 수 있게 합니다.

도구는 에이전트를 순수한 대화 시스템에서 실행 가능하고 생산적인 비서로 변화시킵니다. 몇 가지 예시는 다음과 같습니다:

- 데이터베이스 쿼리 실행 및 특정 정보 검색
- 복잡한 계산 및 통계 분석 수행
- 외부 APIs 및 웹 서비스와 상호작용
- 다양한 형식으로 데이터를 처리 및 변환
- 보고서, 시각화 및 요약 생성
- 현재 데이터를 기반으로 실시간 의사결정

### B2. 도구가 전통적인 AI 컴포넌트와 어떻게 다른가요

에이전트 도구가 다른 AI 기술과 어떻게 연계되는지 이해하는 것이 중요합니다. 다음은 도구와 머신 러닝 모델, 챗봇, 전통적인 APIs를 구분하는 데 도움이 되는 예시입니다:

#### **도구 vs. 머신 러닝 모델**
- **ML 모델**: 에이전트가 사용하는 정보(예측, 생성, 추론)를 제공합니다
- **에이전트 도구**: 에이전트가 행동을 취하거나 정보를 가져오기 위해 호출할 수 있는 실행 가능한 기능입니다. 일부 도구는 머신러닝 모델을 호출할 수도 있고, 다른 도구는 API나 데이터베이스를 호출하거나 비즈니스 로직을 실행할 수 있습니다.
- **예시**: 감정 분석 모델이 고객 메시지의 점수를 매기면, 에이전트는 그 점수를 바탕으로 조치를 취하기 위해 도구(예: `escalate_ticket` )를 사용합니다.

#### **도구 vs. 챗봇**
- **챗봇**: 제한된 범위(스크립트, 검색, 미리 정의된 흐름) 내에서 대화형 답변을 제공합니다.
- **에이전트 도구**: 에이전트가 단순히 답변하는 수준을 넘어설 수 있도록 해줍니다. 에이전트는 액션을 실행하기로 결정할 수 있습니다(예: 데이터베이스 검색, 이메일 발송, 레코드 기록 등).
- **핵심 포인트**: 챗봇은 대화를 나누고, 에이전트는 현실 세계에서 *무언가를 수행*하기 위해 툴을 사용합니다.

#### **도구 vs. 기존 APIs**
- **기존 APIs**: 함수 선택 및 호출을 위해 수동 프로그래밍이 필요합니다
- **에이전트 도구**: 맥락과 목표에 따라 AI의 추론을 통해 동적으로 선택되고 조율될 수 있습니다.
- **인텔리전스**: 도구는 메타데이터와 설명을 제공하여 에이전트가 *언제*와 *어떻게* 사용할지 이해할 수 있도록 합니다.

### B3. 도구 선택 및 오케스트레이션

현대 AI 에이전트의 핵심 역량 중 하나는 지능적인 도구 선택입니다. 사용자의 요청을 받으면 에이전트는 다음과 같은 작업을 수행해야 합니다.

1. **요청 분석**: 사용자가 달성하려는 목적이 무엇인지 이해합니다.
2. **필요한 도구 식별**: 요청을 해결하는 데 어떤 도구가 필요한지 판단합니다.
3. **실행 순서 계획**: 도구를 호출할 순서를 결정합니다.
4. **실행 및 조정**: 적절한 매개변수를 사용하여 도구를 호출하고 그 응답을 처리합니다.
5. **결과 취합**: 여러 도구에서 얻은 결과물을 일관성 있는 하나의 답변으로 통합합니다.
6. **학습 및 적응**: 성공 패턴을 바탕으로 도구 선택 능력을 향상시킵니다.

이러한 조율(오케스트레이션) 역량 덕분에 에이전트는 복잡하고 여러 단계로 이루어진 워크플로우를 자동으로 처리할 수 있으며, 동적인 문제 해결 방식이 필요한 시나리오에 유연하게 대처할 수 있습니다.

## C. Unity Catalog 그리고 에이전트 도구 거버넌스
자신에게 가장 적합한 도구 활용 사례를 결정하려면 Databricks의 도구 생태계를 이해하는 것이 중요합니다. 현재 에이전트 도구를 만드는 방법은 세 가지가 있습니다:
1. **Unity Catalog 함수 도구**: 이 과정에서 주로 다루는 핵심 내용입니다. 도구가 UC UDF(사용자 정의 함수)로 정의되며, 에이전트 도구의 중앙 레지스트리로서 Unity Catalog에서 관리됩니다. 이를 통해 내장된 보안 및 컴플라이언스 기능을 활용할 수 있을 뿐만 아니라, 도구를 더 쉽게 발견하고 재사용할 수 있습니다.
1. **에이전트 코드 도구**: 에이전트의 코드 내에 직접 정의되는 도구입니다. REST API를 호출하거나, 임의의 코드를 실행하거나, 지연 시간이 짧은(low-latency) 도구를 실행하는 데 가장 적합합니다. 다만, 이 방식은 Unity Catalog가 제공하는 내장 거버넌스 및 발견 용이성 기능이 부족합니다.
1. **모델 컨텍스트 프로토콜(MCP) 도구**: 도구 간 상호 운용성을 위한 MCP 표준을 따르는 도구입니다. 현재 Databricks에서 관리하는 MCP 서버를 사용할 수 있으며, 출시 상태는 [여기](https://docs.databricks.com/aws/en/generative-ai/mcp/managed-mcp)에서 확인할 수 있습니다.

### C1. 에이전트 도구로 왜 유니티 카탈로그(Unity Catalog)를 사용해야 할까요?
지금까지 에이전트를 에이전트답게 만드는 기본 요소들을 살펴보았습니다. 이제부터는 데이터브릭스(Databricks)가 도구 호출(Tool Calling)을 어떻게 지원하는지 이해하기 위해, 플랫폼 내에서 도구가 어디에 저장되고 유니티 카탈로그를 통해 어떻게 관리되는지 알아보겠습니다.

![example-agent-framework-with-uc.png](../Includes/images/example-agent-framework-with-uc.png "example-agent-framework-with-uc.png")

<p align="center"><em>전통적인 도구 호출은 포괄적인 거버넌스가 부족합니다. Unity Catalog를 통해 사용자는 구조화 및 비구조화 데이터를 검색하는 도구를 만들고 AI Playground에서 해당 도구를 테스트할 수 있습니다. 외부 도구(예: Slack, Google 캘린더, 또는 API 서비스)를 Unity Catalog 연결을 통해 연결할 때, 자격 증명과 인증 관리는 Unity Catalog 정책에 의해 관리됩니다. 이는 안전하고 감사 가능한 접근을 보장하고, 외부 서비스와의 통합에 대해 조직 전체의 거버넌스를 적용할 수 있음을 의미합니다 </em></p>

**Unity Catalog**는 엔터프라이즈급 기능을 바탕으로 에이전트 도구 관리의 기반을 제공합니다.

1. **중앙 집중식 거버넌스(Centralized Governance)**
    - 유니티 카탈로그가 활성화된 모든 워크스페이스에서 함수를 포함한 데이터 및 AI 자산에 대해 통일된 객체 모델과 3단계 네임스페이스를 제공합니다.
    - 자체 내장된 감사(Auditing) 및 데이터 계보(Lineage) 추적 기능을 제공하며, 시스템 테이블을 통해 액세스 확인과 분석을 단순화합니다.
    - 카탈로그 탐색기(Catalog Explorer) 및 검색 기능을 통해 일관된 메타데이터와 자산 검색 편의성을 제공합니다.
    - 관리형 도구 지원: 유니티 카탈로그에 등록된 함수를 에이전트 도구로 활용할 수 있어, 재사용성과 통제력을 높입니다.
1. **보안 및 액세스 제어(Security and Access Control)**
    - 함수 및 도구에 대한 EXECUTE 권한을 포함하여 세부적인 권한 제어(ANSI GRANT)가 가능합니다.
    - 워크스페이스 간 일관된 액세스를 위해 중앙 집중식 ID 통합(SCIM, 계정 수준 ID)을 지원합니다.
    - Python UDF(사용자 정의 함수)의 안전하고 격리된 실행을 보장하며, 외부 연결 및 위치에 대한 관리형 액세스를 제공합니다.
        - Python UDF를 사용하려면 유니티 카탈로그와 서버리스/Pro SQL 웨어하우스 또는 유니티 카탈로그가 활성화된 클러스터가 필요합니다.
    - 카탈로그 → 스키마 → 객체(테이블, 뷰, 볼륨, 모델, 함수)로 이어지는 역할 기반의 계층적 권한 구조를 따릅니다.
1. **검색 편의성 및 문서화(Discoverability and Documentation)**
    - 풍부한 메타데이터(함수 및 매개변수 주석), 데이터 계보, 브라우징 기능을 갖춘 검색 가능한 카탈로그를 제공합니다.
    - 도구 호출의 정확성을 높이기 위해 함수에 권장 독스트링(Docstring: 목적, 매개변수, 반환 값, 예시, 변경 이력 등) 작성을 지원합니다.
    - 관리형 자산을 더 빠르게 찾을 수 있도록 플랫폼 차원에서 AI 기반 문서화 기능을 지원합니다.
1. **확장성 및 성능(Scalability and Performance)**
    - 유니티 카탈로그의 관리를 받는 도구들은 데이터브릭스 컴퓨팅 자원을 통해 실행됩니다. 에이전트 도구 실행에는 서버리스 범용 컴퓨팅(Spark Connect 서버리스)이 사용됩니다. 일부 연동 환경에서는 SQL 웨어하우스(uc_function)를 통해 유니티 카탈로그 함수를 실행할 수 있습니다.
    - SQL 웨어하우스의 확장 및 동시성 제어 기능과 워크로드 요구량에 맞춘 클러스터 자동 확장(Autoscaling) 기능을 제공합니다.
1. **외부 도구 지원(External Tool Support)**
    - [외부 도구](https://docs.databricks.com/aws/en/generative-ai/agent-framework/external-connection-tools)(예: Slack, Google 캘린더, 또는 모든 API 서비스)를 Unity Catalog 연결을 할 때,자격 증명 및 인증 관리는 유니티 카탈로그 정책의 통제를 받습니다. 이를 통해 외부 서비스 연동 시 안전하고 감사 가능한 액세스를 보장하며, 조직 전반의 거버넌스를 적용할 수 있습니다.

### C2. 도구 등록 및 관리

Unity Catalog은 SQL 기반 에이전트 도구를 등록하고 관리할 수 있는 다양한 방법을 제공합니다. 에이전트의 도구 사용을 추적하기 위해, Databricks는 자동 시그니처 및 추적 기능, 그리고 ResponseAgent 인터페이스와 같은 관리형 MLflow 에이전트 도구 기능을 활용합니다. 본 과정에서는 여러분이 에이전트 개발을 위한 Databricks 플랫폼 사용법에 집중할 수 있도록, 도구 로직을 단순하고 직관적으로 유지하면서 도구 호출(Tool Calling)의 기본 핵심에 초점을 맞출 것입니다. (예: 에이전트의 [벡터 검색](https://www.databricks.com/product/machine-learning/vector-search) 기능 등은 다루지 않습니다.)

#### **SQL 기반 등록**
LLM과 함께 사용할 수 있도록 포괄적인 메타데이터가 포함된 `CREATE OR REPLACE FUNCTION` 문을 사용합니다.
- 타입과 설명이 명확하게 정의된 매개변수
- 함수 수준의 문서화 및 사용 지침
- 결정론적(Deterministic) 동작 사양 정의
- 내장된 유효성 검사 및 오류 처리

#### **프로그래밍 방식 등록**
자동화된 도구 관리를 위해 `DatabricksFunctionClient()` 를 사용합니다.
- 프로그래밍 방식의 생성 및 업데이트
- CI/CD 파이프라인과의 연동
- 일괄 작업 및 대량 관리
- 자동화된 테스트 및 유효성 검사 워크플로우

#### **문서화 모범 사례**
AI 에이전트가 함수의 목적을 잘 이해할 수 있도록 SQL 함수에 풍부한 메타데이터를 포함해야 합니다.
- 비즈니스 로직을 설명하는 상세한 함수 주석
- 예상 데이터 타입 및 범위가 포함된 매개변수 설명
- 반환 값 사양 및 예시 출력
- 사용 예시 및 공통 패턴
- [적절한 경우](https://docs.databricks.com/aws/en/sql/language-manual/sql-ref-syntax-ddl-create-sql-function#parameters) 함수를`DETERMINISTIC`(결정론적)으로 표시

두 가지 접근 방식 모두 거버넌스 및 보안 표준을 유지하면서, SQL 함수가 올바르게 문서화 및 버전 관리되고 AI 에이전트가 접근할 수 있도록 보장합니다.

MLflow는 Databricks에서 에이전트 기반 애플리케이션을 구축, 모니터링, 배포하기 위한 기반이 되며, 특히 에이전트 도구를 사용할 때 강력한 추적, 버전 관리, 평가 및 운영 배포 기능을 제공합니다.

### C3. 기타 도구와 일반적인 패턴
이 과정에서는 UC(Unity Catalog)의 에이전트 도구에 초점을 맞추지만, 이번 과정에서 다루는 것 외에도 다양한 도구가 존재한다는 점을 짚고 넘어갈 필요가 있습니다.

#### 모델 컨텍스트 프로토콜 (MCP)
MCP의 가장 큰 장점은 표준화입니다. 도구를 한 번만 개발하면 직접 구축한 에이전트든 서드파티 에이전트든 관계없이 어떤 에이전트와도 함께 사용할 수 있습니다. 마찬가지로 사내 팀이나 외부 조직 등 다른 사람이 개발한 도구도 그대로 활용할 수 있습니다.
> Databricks의 MCP에 관한 자세한 내용은 [여기](https://learn.microsoft.com/ko-kr/azure/databricks/generative-ai/mcp/)에서 확인할 수 있습니다. 공식 MCP 문서도 [여기](https://modelcontextprotocol.io/docs/getting-started/intro)에서 읽어보실 수 있습니다

#### Databricks AI Search
Databricks AI 검색은 Databricks 데이터 인텔리전스 플랫폼에 내장된 벡터 검색 솔루션으로, 플랫폼의 거버넌스 및 생산성 도구와 통합되어 있습니다. 벡터 검색은 임베딩 추출에 최적화된 검색 방식입니다.
> 벡터 검색에 대해 더 알아보려면 [여기](https://learn.microsoft.com/ko-kr/azure/databricks/ai-search/ai-search)를 참조하세요.

## 결론

이제 여러분은 Databricks 기반의 AI 에이전트와 UC(Unity Catalog) 함수 도구를 뒷받침하는 개념과 원리에 대해 포괄적인 토대를 갖추게 되었습니다. 이번 강의에서는 단순한 규칙 기반 시스템에서 출발하여, 오늘날 전 세계 다양한 산업을 변화시키고 있는 도구 기반의 정교한 시스템에 이르기까지 AI 에이전트의 발전 과정을 살펴보았습니다.

이 강의의 주요 내용은 다음과 같습니다:

- **AI 에이전트**는 복잡한 문제를 해결하기 위해 인지, 의사결정, 행동 역량을 결합한 자율형 시스템입니다.
- **에이전트 도구**는 외부 시스템, 데이터 소스, 특화된 기능과의 인터페이스를 제공함으로써 AI의 역량을 확장합니다.
- **Unity Catalog**는 기업용 에이전트 도구를 배포하는 데 필요한 거버넌스, 보안 및 관리 프레임워크를 제공합니다.

이제 에이전트가 무엇인지, 도구가 에이전트 행동의 핵심 요소임을 이해했으니, Databricks에서 Unity Catalog 도구에 대해 좀 더 깊이 살펴보겠습니다.
