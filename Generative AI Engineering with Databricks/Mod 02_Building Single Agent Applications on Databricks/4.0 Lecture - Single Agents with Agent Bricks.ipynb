# 강의 - Agent Bricks를 활용한 단일 에이전트 구축

## 개요

Agent Bricks는 기술 사용자가 고품질의 도메인 특화형 AI 에이전트를 상용 서비스 수준으로 신속하게 구축하고 최적화할 수 있도록 지원하는 고수준 추상화 플랫폼입니다. 특히, 비용 부담을 최소화하면서도 에이전트의 품질을 극대화할 수 있도록 인간 피드백 기반 에이전트 학습(ALHF)을 포함한 자동 평가 및 최적화 기능에 초점을 맞추고 있습니다.

수많은 수동 설정과 최적화 과정을 거쳐야 하는 기존의 에이전트 개발 방식과 달리, **Agent Bricks는 구현 프로세스를 대폭 간소화** 합니다. 이를 통해 사용자는 하위 수준의 기술적 세부 사항에 얽매이지 않고 문제 해결, 데이터, 그리고 핵심 지표에만 집중할 수 있습니다. 또한, 이 플랫폼은 각각 특정 유스케이스와 배포 패턴에 맞춤화된 네 가지 고유한 에이전트 유형을 지원합니다.

## 학습 목표
_본 강의를 마치면 여러분은 다음 역량을 갖추게 됩니다._

- Agent Bricks의 개발 라이프사이클과 반복적 최적화 프로세스 이해
- 지원되는 네 가지 에이전트 유형과 각 유형의 구체적인 사용 사례 파악
- 자동화형(Automated) 에이전트와 대화형(Interactive) 에이전트 분류의 차이점 설명
- 비용-성능 균형을 위한 최적화 전략 서술
- Agent Bricks에 내장된 평가 및 모니터링 기능 파악

## A. Agent Bricks 소개

![agent-bricks-cost.png](../Includes/images/agent-bricks-cost.png "agent-bricks-cost.png")

Agent Bricks는 특정 도메인에 특화된 에이전트 시스템을 구축할 수 있는 단순하면서도 강력한 방법을 제공합니다. 본 플랫폼은 기업용 애플리케이션에 필수적인 유연성을 유지하면서도, 기존의 에이전트 개발 과정에서 수반되던 복잡성을 상당 부분 제거해 줍니다.

Agent Bricks의 핵심 철학은 사용자가 비즈니스 문제를 정의하고 관련 데이터를 제공하는 데만 집중할 수 있도록 하는 것입니다. 에이전트 최적화, 평가, 배포 등 기술적인 복잡성은 플랫폼이 대신 처리합니다. 이러한 방식을 통해 기업 환경에서 AI 에이전트를 도입하고 가치를 창출하기까지 걸리는 시간(Time-to-Value)을 크게 단축할 수 있습니다.

### A1. 지원되는 에이전트 유형 및 사용 사례

Agent Bricks는 특정 기업의 활용 사례와 운영 패턴에 맞춤 설계된 네 가지 주요 에이전트 유형을 지원합니다. 귀사의 구체적인 요구사항에 맞는 적절한 에이전트 유형을 선택하기 위해서는 이러한 차이점을 이해하는 것이 매우 중요합니다.

**네 가지 에이전트 유형:**

1. **정보 추출(IE)**: 문서, PDF, 이메일, 이미지 등 비정형 소스에서 구조화된 데이터를 자동으로 추출합니다.
2. **맞춤형 LLM (CLLM)**: 특정 작업과 데이터셋에 맞춰 미세 조정(Fine-tuning) 및 최적화된 도메인 특화 대형 언어 모델입니다.
3. **지식 어시스턴트(KA)**: 검색 증강 생성(RAG) 기술을 활용하여 지식 베이스를 기반으로 질의응답 기능을 제공하는 대화형 에이전트입니다. 즉, KA는 도구 호출(Tool-calling) 기능이 RAG 애플리케이션으로만 제한된 단일 에이전트입니다.
4. **다중 에이전트 관리자(MAS)**: 복잡하고 여러 단계로 이루어진 작업을 완료하기 위해 특화된 여러 에이전트를 관리하고 조율하는 조정 시스템입니다. 예를 들어, MAS에 추가 에이전트 없이 일련의 도구 모음만 부여하여 툴킷을 갖춘 단일 에이전트처럼 작동하게 만들 수도 있습니다.

### 지니 에이전트 (Genie Agent)
사용자는 지니 에이전트를 생성 및 활용하여 자연어로 데이터베이스나 기타 구조화된 데이터를 조회할 수 있으며, 이를 통해 데이터 분석을 더욱 쉽게 진행할 수 있습니다. 지니 에이전트는 MAS와 함께 조율되어 작동하거나 단독 에이전트로 구동될 수 있습니다.

### A2. 운영 분류


![automated-interactive-agents.png](../Includes/images/automated-interactive-agents.png "automated-interactive-agents.png")

에이전트는 의도된 사용 패턴에 따라 두 가지 운영 모델로 분류됩니다.

- **자동화 브릭(Information Extraction 및 Custom LLM)**: 인간의 개입을 최소화한 채 대규모 배치(일괄) 처리를 수행하는 시나리오에 최적화되어 있습니다. 이 에이전트들은 가성비(비용 대비 성능) 최적화와 처리량 향상에 중점을 둡니다.
- **대화형 브릭(Knowledge Assistant, Multi-Agent Supervisor 및 Genie):** 인간이 개입하는(Human-in-the-loop) 경험과 실시간 상호작용 시나리오를 위해 설계되었습니다. 이 에이전트들은 대화형 인터페이스와 동적인 답변 생성에 중점을 둡니다.

## B. Agent Bricks 개발 수명 주기

에이전트 브릭스의 개발 프로세스는 지속적인 개선과 피드백 반영을 통해 에이전트의 성능을 최적화하도록 설계된 체계적이고 반복적인 접근 방식을 따릅니다. 이 라이프사이클을 통해 에이전트가 초기 요구사항을 충족할 뿐만 아니라, 실제 사용 경험과 피드백을 바탕으로 지속적으로 발전할 수 있도록 보장합니다.

### B1. 핵심 3단계 개발 주기



에이전트 브릭스 개발 라이프사이클은 에이전트 개발의 기반이 되는 세 가지 주요 단계와 지속적인 개선을 위한 반복 과정으로 구성됩니다.

**1단계: 문제 정의(Specify Your Problem)**
![agent-bricks-high-level-architecture.png](../Includes/images/agent-bricks-high-level-architecture.png "agent-bricks-high-level-architecture.png")

대략적인 흐름을 살펴보면, 사용자는 먼저 자신의 특정 사용 사례에 맞는 에이전트를 구축하는 것부터 시작합니다. 예를 들어, MAS(멀티 에이전트 감독)를 사용하는 경우 Genie 에이전트와 함께 도구 호출(Tool Calling)만 허용하는 관리형 에이전트를 원할 수 있습니다. 적절한 권한을 설정한 후에는 MLflow를 활용하여 메트릭을 추적하고 로그를 기록합니다.

이 초기 단계에서는 AI 에이전트의 범위와 요구사항을 다음과 같이 정의합니다.
- 팀과 함께 필요한 작업과 예상 결과물을 명확하게 정의합니다.
- 제공되는 네 가지 옵션(정보 추출, 맞춤형 LLM, 지식 어시스턴트, 멀티 에이전트 감독) 중에서 적절한 에이전트 유형을 선택합니다.
- 사용 사례에 따라 UC(Unity Catalog) 관리형 데이터 세트(Delta 테이블, UC 볼륨)를 제공하고, 도구를 장착하며, 다른 에이전트를 연결해야 합니다.
- 평가를 위한 성공 기준과 품질 메트릭을 수립합니다.

**2단계: 기업 데이터 기반 최적화 (Optimize on Your Enterprise Data)**

에이전트 브릭스는 품질과 비용 간의 절충안을 고려하여 최적의 에이전트 시스템을 자동으로 구축하고 최적화합니다.
- 시스템이 특정 작업과 관련된 평가 벤치마크(정확도, 제품 관련성, 고객 이탈 예측 등)를 자동으로 생성합니다.
- 최적화 과정에서는 검증된 여러 기술을 지능적으로 선택하고 조합합니다.
  - 검증된 방법론을 활용한 고급 프롬프트 최적화
  - 작업 요구사항 및 데이터 가용성에 기반한 선택적 파인튜닝(미세조정)
  - 최적의 도구 선택 및 구성
  - 품질 평가를 위한 맞춤형 LLM 평가자(Judge) 구현
  - 응답 품질 향상을 위한 리워드 모델(Reward Model) 필터링
  - 효과가 있는 경우, 인간 피드백 기반 강화학습(RLHF) 적용


![setup-architecture-1.png](../Includes/images/setup-architecture-1.png "setup-architecture-1.png")


**3단계: 지속적인 개선 (Continuous Improvement)**

마지막 단계에서는 지속적인 최적화를 위한 피드백 루프를 구축합니다.
- 최적화된 에이전트를 프로덕션 환경에 배포합니다.
- 자동화된 평가와 사람에 의한 평가를 통해 에이전트의 품질을 지속적으로 측정합니다.
- 모니터링을 통해 문제점과 개선 기회를 체계적으로 식별합니다.
- 시스템 성능을 향상하기 위해 자연어 피드백을 적용합니다.
- 반복적인 성능 향상을 위해 인간 피드백 기반 에이전트 학습(ALHF)을 활용합니다.

### B2. 평가 및 모니터링 프레임워크

에이전트 브릭스는 플랫폼에 직접 내장된 포괄적인 평가 및 모니터링 기능을 제공하여, 에이전트의 성능과 품질 메트릭을 지속적으로 시각화하고 파악할 수 있도록 지원합니다.

**자동 MLflow 통합:**

에이전트 브릭스를 통해 배포되는 모든 에이전트에는 다음과 같은 포괄적인 추적 기능이 자동으로 포함됩니다.
- **요청 추적**: 타임스탬프 및 사용자 컨텍스트를 포함한 모든 인입 요청의 완전한 로그 기록
- **응답 모니터링**: 신뢰도 점수 및 추론 경로를 포함한 출력 응답의 상세한 캡처
- **에이전트 간 통신**: 다중 에이전트 시스템에서 에이전트 간에 이루어지는 통신의 전체 트레이싱(추적)
- **성능 지표**: 대기 시간(Latency), 처리량(Throughput), 리소스 데이터의 자동 수집

**품질 평가 메커니즘:**

플랫폼은 다각적인 품질 평가 레이어를 구현합니다.
- **자동 벤치마크 생성**: 사용자의 요구사항에 맞춘 작업별 메트릭 생성
- **LLM 평가자(Judge) 검증**: 평가 작업에 특화되어 학습된 전용 언어 모델을 활용한 자동 품질 점수 측정
- **인간 피드백 통합**: 리뷰 애플리케이션을 통해 전문가의 피드백을 체계적으로 수집하고 반영
- **프로덕션 성능 모니터링**: 실제 라이브 환경에서 에이전트 성능의 실시간 추적
- **비교 분석**: 기준(Baseline) 모델 및 이전 에이전트 버전과의 벤치마킹 비교

## C. 다른 서비스와의 통합
![agent-bricks-integration.png](../Includes/images/agent-bricks-integration.png "agent-bricks-integration.png")

Agent Bricks는 Mosaic AI Model Serving, AI search, Unity Catalog, Genie, MLflow 3, Databricks Apps와 긴밀히 통합되어 **AI 에이전트를 엔드투엔드에서 구축, 관리, 평가, 배포하는 통합 플랫폼을 제공합니다.** 이러한 연동 덕분에 사용자는 업계 최고 수준의 거버넌스, 보안 및 확장성을 유지하면서 기업 내부 데이터를 활용해 에이전트 시스템을 신속하게 프로토타입화하고 반복 개선하며 배포할 수 있습니다.

#### Agent Bricks가 Databricks 스택의 유기적 작동 방식

- **Mosaic AI Model Serving**: 에이전트를 자동 부하 분산 및 모니터링 기능이 포함된 확장 가능한 REST API로 배포할 수 있습니다. 이 서빙 플랫폼은 안전한 인증을 제공하며 MLflow 3와 기본적으로 연동되어 실시간 트레이싱과 품질 평가를 지원합니다.
- **AI Search**: 에이전트는 Databricks AI Search를 활용해 관련된 비정형 정보를 효율적으로 검색합니다. 이는 검색 증강 생성(RAG)뿐만 아니라 문서 및 테이블 전체에 대한 시맨틱 검색과 같은 고급 유스케이스도 지원합니다.
- **Unity Catalog**: 모든 데이터, 모델, 에이전트 및 툴에 걸쳐 통합된 거버넌스를 보장합니다. 규제 및 컴플라이언스 요구사항을 충족하기 위해 에이전트 로직, 데이터 계보(lineage), 툴 접근 권한을 제어하는 동시에 기업 보안 요구사항과의 연동을 지원합니다.
- **Genie & Genie Spaces**: 에이전트가 정형 데이터와 직접 상호작용(예: Text-to-SQL 쿼리)하고 다양한 툴을 조율할 수 있도록 지원하여, 에이전트의 역량(예: 멀티 에이전트 또는 툴 호출 아키텍처)을 확장합니다.
- **MLflow 3**: 에이전트를 위한 강력한 실험 추적, 버전 관리, 트레이싱 및 평가 기능을 제공합니다. 실시간 트레이스 분석과 자동화된 품질 측정(연구 기반의 LLM 평가자 활용)을 통해 신속한 디버깅과 개선 사이클이 가능해집니다.
- **Databricks Apps**: 빌트인 채팅 앱, 피드백 수집 포털, 운영 대시보드 등의 사용자 인터페이스를 제공합니다. 이해관계자들은 이러한 UI를 통해 에이전트와 상호작용하고 피드백을 제출하며, 에이전트가 비즈니스 요구사항을 충족하는지 확인할 수 있습니다.
