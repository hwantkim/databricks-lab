# 강의 - 모범 사례와 실무 적용

## 개요

이번 마지막 강의에서는 에이전트 평가를 위한 모범 사례를 종합하고, 실무 적용을 위한 가이드를 제공합니다. 종합적인 평가를 위한 팁을 살펴보고, MLflow 에코시스템 연동을 이해하며, 이러한 개념들을 실제 비즈니스 시나리오에 적용하는 방법을 알아봅니다.

본 강의에서는 평가가 처한 상황과 맥락에 따라 크게 달라진다는 점을 강조하며, 여러분의 애플리케이션, 사용자, 그리고 비즈니스 요구사항에 특화된 평가 요건을 고려할 수 있는 프레임워크를 제공합니다.

**학습 목표**

본 강의를 마치면 여러분은 다음을 수행할 수 있게 됩니다.
- 종합적인 에이전트 평가를 위한 모범 사례 적용
- MLflow가 더 넓은 평가 에코시스템과 연동되는 방식 이해
- 각자의 사용 사례에 맞는 평가 전략을 설계하기 위한 핵심 질문 식별
- 평가를 일회성 활동이 아닌 지속적인 규율(Discipline)로 인식
- 지속적인 개선을 지원하는 평가 인프라 구축 계획 수립

## A. 에이전트 평가를 위한 팁

### A1. 종합 평가 전략

![tips-for-agent-evaluation.png](../Includes/images/Evaluation with MLflow/tips-for-agent-evaluation.png "tips-for-agent-evaluation.png")

효과적인 에이전트 평가를 위해서는 품질의 다각적 측면, 다양한 평가 시나리오, 지속적 개선을 위한 체계적인 프로세스를 모두 고려하는 총체적인 접근 방식이 필요합니다.

### A2. 다차원 품질 평가

**고려해야 할 품질 요인**

- **정확성:** 답변에 포함된 사실관계가 정확한가?
- **연관성:** 답변이 사용자의 실제 질문에 부합하는가?
- **완결성:** 답변이 장황하지 않으면서도 핵심을 모두 담고 있는가?
- **안전성:** 답변에 유해하거나 부적절한 콘텐츠가 없는가?
- **일관성:** 유사한 질문에 대해 일관된 품질의 답변이 제공되는가?
- **효율성:** 에이전트가 도구와 리소스를 적절하게 사용하는가?
- **사용자 경험:** 답변이 유용하고 명확하며 보기 좋게 구성되어 있는가?

### A3. 평가 데이터셋 구축 모범 사례

**견고한 평가 데이터셋 구축**

1. **사용자 조사부터 시작:** 실제 사용자 질문과 페인 포인트(불편 사항)를 분석합니다.
2. **실패 사례 포함:** 에이전트가 답변을 거부하거나 다른 곳으로 안내해야 하는 시나리오를 테스트합니다.
3. **복잡도 다양화:** 단순한 정보 조회부터 복잡한 다단계 추론까지 골고루 포함합니다.
4. **한계점 테스트:** 예외적인 케이스(Edge case)와 모호한 질문을 포함합니다.
5. **다양성 유지:** 다양한 도메인, 질문 스타일, 사용자 유형을 아우릅니다.
6. **체계적인 버전 관리:** 에이전트 버전과 함께 데이터셋의 변경 사항을 추적합니다.
7. **정기적인 검증:** 정답 데이터(Ground truth)가 여전히 정확하고 유효한지 확인합니다.

### A4. 평가 워크플로우 모범 사례

**체계적 평가 프로세스**

- **가능한 부분은 자동화**: CI/CD 파이프라인을 활용하여 코드 변경 사항에 대한 평가를 수행합니다.
- **품질 기준(Quality Gate) 설정**: 배포를 위한 최소 기준치를 정의합니다.
- **체계적인 비교 분석**: 변경 사항을 항상 기준 성능(Baseline)과 비교하여 평가합니다.
- **결정 근거 문서화**: 평가 결정의 배경이 된 논리 및 근거를 기록으로 남깁니다.
- **결과 공유**: 평가 결과를 이해관계자들이 열람할 수 있도록 공개합니다.
- **인사이트 기반 실행**: 평가 결과를 활용하여 구체적인 개선 조치를 추진합니다.

## B. MLflow 평가 생태계(Evaluation Ecosystem)

### B1. 생태계 통합(Ecosystem Integration)

![mlflow-ecosystem.png](../Includes/images/Evaluation with MLflow/mlflow-ecosystem.png "mlflow-ecosystem.png")

MLflow 평가는 더 넓은 Databricks 생태계와 통합되어, 개발 단계부터 운영 모니터링에 이르기까지 포괄적인 평가 기능을 제공합니다.

### B2. MLflow 실험(Experiments) 및 실행(Runs)

MLflow 실험은 평가 결과를 체계적으로 정리하여, 에이전트의 반복 작업 전반에 걸쳐 비교 및 분석을 가능하게 합니다.

**실험 구조**

`mlflow.genai.evaluate()`를 호출하여 평가를 수행할 때마다 실험 내에 하나의 '실행(run)'이 생성됩니다.
- **실행(Runs):** 특정 구성을 가진 개별 평가 실행을 나타냅니다.
- **실험(Experiments):** 관련된 평가 실행을 비교를 위해 그룹화합니다.
- **메트릭(Metrics):** 실행 수준에서 기록됩니다.(집계된 점수)
- **산출물(Artifacts):** 상세한 예제별 결과 및 평가 데이터셋을 포함합니다.

### B3. 비교 및 분석 모범 사례 (Best Practices)

**비교 기능**

MLflow UI를 통해 체계적인 비교가 가능합니다.
- 실행(Run) 간 지표를 비교하여 개선 사항을 정량화
- 시간 경과에 따른 지표 추세를 시각화
- 태그, 파라미터 또는 지표를 기준으로 실행을 필터링
- 외부 분석을 위해 결과를 내보내기

**모범 사례**

- 일관된 실험 명명 규칙을 사용합니다.
- 의미 있는 메타데이터(에이전트 버전, 구성, 데이터셋 버전)로 실행에 태그를 지정합니다.
- 하이퍼파라미터와 구성을 실행 매개변수로 기록합니다.
- 실행 컨텍스트와 결과를 설명하는 메모를 추가합니다.

## C. 실무 적용 프레임워크

### C1. 유스케이스별 핵심 질문

시연 과정을 살펴보면서 다음 개념들을 여러분의 유스케이스에 어떻게 적용할 수 있을지 고민해 보세요.

**생각해 볼 만한 질문**
- 여러분의 애플리케이션에서 가장 중요한 품질 요소는 무엇인가요?
- 실제 운영 환경에서 어떤 오류 유형이 가장 치명적일까요?
- 실제 사용 현황을 반영한 평가 데이터셋을 어떻게 구축할 것인가요?
- 기존의 어떤 시스템이나 프로세스를 평가 기준으로 삼을 수 있을까요?
- 개발 워크플로우에 평가 프로세스를 어떻게 통합할 것인가요?
- 운영 모니터링 시 어떤 지표들을 추적할 것인가요?

### C2. 맥락 의존적 평가

효과적인 평가는 철저히 맥락에 의존합니다. MLflow가 강력한 도구들을 제공하지만, 여러분의 특정 애플리케이션과 사용자, 그리고 비즈니스 요구사항에 맞는 '올바른 기준'은 직접 정의해야 합니다.

**고려해야 할 맥락**

- **도메인 요구사항:** 의료, 법률, 금융 분야의 애플리케이션은 특화된 평가 기준이 필요할 수 있습니다.
- **사용자 기대치:** 전문가 사용자는 일반 소비자와 다른 형태의 답변 스타일을 기대할 수 있습니다.
- **리스크 허용도:** 중요도가 높은 애플리케이션은 실험적 도구보다 훨씬 더 엄격한 평가가 필요합니다.
- **자원 제약:** 평가의 철저함과 개발 속도 사이에서 균형을 맞춰야 합니다.
- **규제 요구사항:** 일부 도메인은 특정 컴플라이언스나 감사 기능을 요구합니다.

### C3. 구현 계획 수립

**평가 실행 계획**

1. **간단하게 시작하기:** 기본적인 기본 제공 평가자(built-in judges)로 시작해 점진적으로 확장하세요.
2. **이해관계자 식별:** 도메인 전문가, 사용자, 비즈니스 담당자를 포함하세요.
3. **성공 지표 정의:** 명확하고 측정 가능한 품질 표준을 수립하세요.
4. **인프라 계획:** 스토리지, 컴퓨팅, 거버넌스 요구사항을 고려하세요.
5. **피드백 루프 설계:** 평가 결과를 개발 프로세스와 연계하세요.
6. **규모 확장 대비:** 평가 인프라가 수요 증가에 맞춰 확장될 수 있도록 준비하세요.

## D. 지속적인 과업으로서의 평가

### D1. 지속적 개선 마인드셋

평가는 일회성 활동이 아니라 계속 다듬어 가야 하는 작업이라는 점을 기억하세요. 에이전트가 진화하고, 사용 패턴이 변하며, 이해가 깊어짐에 따라 평가 방식도 진화해야 합니다. MLflow의 유연한 프레임워크는 엄격성과 재현성을 유지하면서 이러한 지속적인 개선을 지원합니다.

**진화의 원동력**

- **에이전트 기능**: 새로운 기능은 새로운 평가 기준을 필요로 합니다.
- **사용자 행동**: 변화하는 사용 패턴은 데이터셋 업데이트가 필요합니다.
- **비즈니스 요구사항**: 우선순위가 바뀌면 품질 정의가 바뀔 수 있습니다.
- **기술적 발전**: 새로운 평가 기법을 활용할 수 있게 됩니다.
- **학습된 교훈**: 프로덕션 환경에서의 경험을 바탕으로 평가를 개선합니다.

### D2. 평가 문화 구축

**우수한 평가 문화 육성**

- **평가 시각화:** 팀 전체에 결과와 인사이트를 공유합니다.
- **개선사항 축하:** 평가를 통해 달성한 품질 향상을 격려합니다.
- **실패로부터 배우기:** 평가 실패를 학습의 기회로 삼습니다.
- **도구에 투자:** 팀에 훌륭한 평가 인프라를 제공합니다.
- **팀원 교육:** 모든 구성원이 평가 원칙과 실무를 이해하도록 돕습니다.

## E. 앞으로의 전망

### E1. 품질을 위한 기반

AI 에이전트의 품질은 결국 평가의 품질에 좌우됩니다. 초기부터 탄탄한 평가 인프라 구축에 투자한다면, 개발부터 실운영 그리고 그 이후에 이르기까지 에이전트의 전체 수명 주기 동안 그 결실을 보게 될 것입니다.

**좋은 평가의 장기적 이점**

- **빠른 개발:** 신속한 피드백을 통해 빠른 반복 개발이 가능해집니다.
- **더 높은 품질:** 체계적인 평가로 문제를 조기에 발견합니다.
- **사용자 신뢰:** 일관된 품질을 통해 사용자의 신뢰를 쌓습니다.
- **운영 효율성:** 자동화된 평가로 수동 테스트를 줄입니다.
- **지속적인 개선:** 데이터 기반의 최적화가 가능해집니다.

### E2. 다음 단계

이제 체계적인 에이전트 평가를 구현하기 위한 개념적 기반이 마련되었습니다. 이어지는 데모를 통해 이러한 개념을 실무 기술로 전환하고, 실제 에이전트를 평가하고 결과를 해석하며 품질 개선을 이끄는 구체적인 방법을 배우게 됩니다.

**앞으로 나아갈 평가 여정**

1. **개념 적용하기:** 실습 데모를 하나씩 해결해 나갑니다.
2. **맥락에 맞추기:** 구체적인 요구사항과 제약 조건을 고려합니다.
3. **시작은 간단하게:** 기본 평가부터 구현하고 점진적으로 확장합니다.
4. **측정 및 개선:** 평가 결과를 활용해 실질적인 개선을 이끌어냅니다.
5. **공유 및 학습:** 평가 역량을 구축 중인 다른 사람들과 협업합니다.

## F. 결론

효과적인 에이전트 평가는 예술과 과학의 조화와 같습니다. 이를 위해서는 평가 프레임워크에 대한 기술적 이해와 품질 기준을 정의할 수 있는 도메인 전문 지식, 그리고 지속적인 개선을 보장하는 체계적인 프로세스가 모두 필요합니다.

MLflow가 기술적 기반을 제공하지만, 성공 여부는 이러한 도구들을 여러분의 구체적인 상황에 맞게 얼마나 신중하게 적용하느냐에 달려 있습니다. 철저한 오프라인 평가와 포괄적인 온라인 모니터링을 결합함으로써, 사용자에게 일관되게 고품질의 경험을 선사하는 에이전트를 구축할 수 있습니다.

평가 인프라에 대한 투자는 에이전트의 전체 수명 주기 동안 큰 결실로 돌아옵니다. 지금 평가 역량을 구축하기 시작한다면, 사용자의 요구를 충족하고 기대를 뛰어넘는 고품질 AI 에이전트를 개발, 배포 및 유지 관리할 수 있는 훌륭한 입지를 확보하게 될 것입니다.
