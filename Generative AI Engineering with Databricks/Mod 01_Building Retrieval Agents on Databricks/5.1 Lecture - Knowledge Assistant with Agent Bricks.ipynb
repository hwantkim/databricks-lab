# Agent Bricks와 함께하는 지식 어시스턴트

## 서론

이 강의에서는 프로덕션 환경에 바로 적용할 수 있는 AI 에이전트를 쉽게 개발할 수 있도록 지원하는 Databricks Mosaic AI의 선언적 프레임워크, **에이전트 브릭스(Agent Bricks)**를 소개합니다. 특히 기업 내부 문서를 기반으로 전문적인 대화형 에이전트를 구축하는 특화된 패턴인 **지식 어시스턴트(Knowledge Assistant)**를 중점적으로 다룹니다. 이번 시간을 통해 에이전트 브릭스가 자동화된 최적화 루프를 활용해 어떻게 개발 패러다임을 수동 튜닝에서 결과 지향적 선언 방식으로 전환하고 효율성을 높이는지 배우게 됩니다. 마지막으로, 이러한 에이전트가 안정성과 확장성, 그리고 거버넌스를 갖출 수 있도록 뒷받침하는 핵심 아키텍처를 함께 살펴봅니다.

## 수업 목표

이 수업이 끝날 때쯤이면 다음과 같은 행동을 할 수 있게 될 것입니다:

* 기존의 수동 개발 방식과 비교하여 에이전트 브릭스(Agent Bricks)가 가진 핵심 가치 제안을 **정의** 할 수 있습니다.  
* 정보 추출 및 맞춤형 대형언어모델(LLM)을 포함하여 에이전트 브릭스의 주요 활용 사례를 **파악** 할 수 있습니다.
* 파싱(구문 분석), 벡터 검색, 모델 서빙 등 지식 어시스턴트(Knowledge Assistant)를 구성하는 아키텍처 요소를 **설명** 할 수 있습니다. 
* '품질 루프(Quality Loop)'의 개념을 설명하고, 인간 피드백 기반 에이전트 학습(ALHF)이 에이전트의 성능을 어떻게 최적화하는지 **기술** 할 수 있습니다.

## A. 에이전트 브릭스(Agent Bricks)란 무엇인가요?

**에이전트 브릭스(Agent Bricks)**는 상용 서비스 수준의 AI 에이전트를 더 빠르게 제작, 배포, 최적화할 수 있도록 지원하는 데이터브릭스 모자이크 AI(Databricks Mosaic AI) 내의 **선언형(declarative)** 프레임워크입니다. 엔지니어가 직접 모델을 고르고, 청킹(chunking) 전략을 설정하며, 프롬프트를 일일이 수정해야 했던 기존의 수동 방식(DIY)과 달리, 에이전트 브릭스는 입력된 데이터와 작업(task)을 기반으로 이러한 구성 결정을 자동화합니다.

### A1. AI 상용화의 과제

생성형 AI를 단순 기술 검증(PoC) 단계에서 실제 상용 서비스로 전환할 때 주로 세 가지 걸림돌에 직면하게 됩니다.

1. **최적화의 복잡성:** AI 시스템에는 LLM 선택(예: Llama 4 vs. GPT-4o), 검색 전략(청크 크기, 임베딩 모델 등), 프롬프트 엔지니어링 기법 등 조정해야 할 '설정 변수(knob)'가 너무나 많습니다. 특정 기업 데이터에 맞는 최적의 조합을 찾아내는 작업은 상당한 시간이 소요됩니다.
2. **평가의 어려움:** 에이전트가 실제 서비스에 투입해도 될 만큼 '충분히 우수한지' 판단하려면 엄격한 테스트가 필요합니다. 하지만 많은 팀이 검증된 정답 데이터(golden datasets)나 신뢰할 수 있는 지표가 부족하여, 직관이나 주관적인 느낌(vibe check)에 의존하곤 합니다.
3. **비용과 품질의 상충 관계(Trade-off):** 고품질을 구현하려면 대개 비용이 많이 드는 대형 모델이 필요하고, 반대로 비용을 줄이려고 하면 성능이 떨어지기 마련입니다. 기업들은 최소한의 비용으로 품질을 극대화하는 최적의 균형점을 찾는 데 애를 먹고 있습니다.

### A2. 에이전트 브릭스가 제시하는 솔루션

에이전트 브릭스는 에이전트 정의를 **선언형 방식**으로 처리하여 이 문제를 해결합니다. 사용자가 데이터와 작업을 지정하기만 하면, 에이전트 브릭스 엔진이 시스템을 반복적으로 알아서 최적화합니다.

이 과정을 이끄는 핵심 메커니즘은 **인간 피드백 기반 에이전트 학습(ALHF, Agent Learning from Human Feedback)**입니다. 시스템은 다음과 같이 작동합니다.

1. 기준 에이전트를 즉시 **배포**합니다.  
2. 리뷰 앱(Review App)을 통해 피드백(좋아요/싫어요, 수정된 답변 등)을 **수집**합니다. 
3. 이 피드백을 종합하여 평가 벤치마크를 자동으로 생성하고, 코드를 수동으로 수정할 필요 없이 내부 프롬프트와 구성을 **최적화** 합니다.

## B. Agent Bricks 사용 사례

Agent Bricks는 일반적인 기업 패턴을 위한 사전 구성된 아키텍처("브릭")를 제공합니다. 각 브릭은 특정 상호작용 및 데이터 처리 방식에 특화되어 있습니다.

### B1. 지식 어시스턴트 (Knowledge Assistant)

이것이 이번 강의의 초점입니다. **지식 어시스턴트는 기업 문서를 전문가의 대화 에이전트로 바꿉니다**.

* **기능:** 지정된 파일을 대상으로 검색 증강 생성(RAG)을 수행합니다. 텍스트 파싱, 청킹(분할), 임베딩 및 인용 출처 생성을 자동으로 처리합니다. 
* **사용 사례:** 사내 규정집을 기반으로 인사 정책 관련 질문에 답하는 HR 봇, 또는 제품 매뉴얼을 바탕으로 기술 문의를 해결하는 기술 지원 봇 등이 있습니다.

### B2. 정보 추출 (Information Extraction)

이 에이전트 유형은 비구조화된 문서(예: PDF, 이미지, 텍스트 파일)를 구조화된 데이터로 변환합니다.

* **기능:** JSON 스키마로 정의된 특정 필드를 추출합니다.
* **사용 사례:** 대량의 인보이스(청구서)를 '공급업체명', '총액', '날짜' 등이 포함된 정형 델타 테이블(Delta table)로 변환하거나, 법률 계약서에서 특정 조항을 추출하는 작업 등이 있습니다.

### B3. 멀티 에이전트 감독자 (Multi-Agent Supervisor)

이 고급 패턴은 여러 에이전트와 도구를 조율하여 복잡하고 다단계적인 문제를 해결합니다.

* **기능:** '감독자(supervisor)' 에이전트가 사용자의 질의를 적절한 하위 에이전트나 도구(예: Unity Catalog Function)로 라우팅(전달)합니다. 
* **사용 사례:** 고객 지원 시스템에서 감독자 에이전트가 결제 관련 문의는 **Genie 스페이스**(정형 데이터)로 보내고, 기술적인 문제 해결 문의는 **지식 어시스턴트**(비정형 데이터)로 전달하는 구조가 있습니다.

### B4. 커스텀 LLM

이 에이전트는 특정 기업 지침과 작업에 맞춘 특화된 LLM 엔드포인트를 생성합니다.

* **기능:** 특정 어조, 서식 또는 준수해야 할 규정을 따르도록 모델을 최적화합니다.
* **사용 사례:** 브랜드의 스타일 가이드를 철저히 준수하여 소셜 미디어 게시물을 작성하는 마케팅 콘텐츠 생성기, 또는 경영진 보고서용 특정 서식으로 결과를 출력하는 요약 도구 등이 있습니다.

## C. 선언적 대 코드 우선 방법

데이터브릭스(Databricks)에서 AI 에이전트를 구축할 때, 개발자는 일반적으로 '코드 우선(Code-First)'과 '선언형(Declarative)'이라는 두 가지 주요 추상화 수준 중 하나를 선택하게 됩니다.

### C1. 코드 우선 방식 (Mosaic AI Agent Framework)

이 방식은 제어 권한을 최대한으로 제공하지만, 그만큼 더 많은 노력이 필요합니다. 개발자는 파이썬(Python) 라이브러리(LangChain, LlamaIndex, OpenAI SDK 등)를 사용해 에이전트의 핵심 로직을 코드로 직접 작성하며, 프레임워크 구축, 트레이싱(추적), 거버넌스를 위해 **Mosaic AI 에이전트 프레임워크**를 활용합니다.

* **워크플로우:** 개발자가 직접 리트리벌(정보 검색) 로직을 작성하고, 프롬프트 템플릿을 정의하며, 임베딩 모델을 선택하고, AI 검색 인덱스 동기화를 관리합니다. 또한 에이전트 프레임워크를 사용해 MLflow에 트레이스를 기록하고, 에이전트를 모델 서빙 엔드포인트로 배포합니다.
* **장점:** 제한 없는 커스터마이징이 가능합니다. 새로운 추론 루프나 매우 구체적인 도구 활용 방식을 직접 구현할 수 있습니다.
* **단점:** 기술 부채를 개발자가 직접 감당해야 합니다. 최적화(청킹, 프롬프트 엔지니어링) 작업이 수동으로 이루어지며, 리트리벌 전략을 변경해야 할 경우(예: 청크 크기 변경) 코드를 다시 작성하여 재배포해야 합니다.

### C2. 선언형 방식 (Agent Bricks)

이 방식은 '결과 중심적' 접근법입니다. 개발자는 에이전트가 '어떻게' 작동해야 하는지가 아니라, '무엇을' 해야 하는지를 선언합니다.

* **워크플로우:** 개발자가 '지식 어시스턴트(Knowledge Assistant)'를 선택하고, PDF가 포함된 유니티 카토로그 볼륨(Unity Catalog Volume)을 지정한 뒤, 페르소나에 대한 텍스트 설명을 제공합니다. 파싱, 인덱싱, 프롬프트 엔지니어링은 Agent Bricks가 알아서 처리합니다.
* **장점:** 가치 창출 시간(Time-to-Value)이 가장 빠릅니다. 시스템이 자체 테스트를 위한 합성 데이터를 생성하고, 피드백을 기반으로 자동 최적화를 수행합니다.
* **단점:** 순수 코드 방식에 비해 하위 수준(low-level)의 실행 로직을 세부적으로 제어하기 어렵습니다.


## D. 지식 어시스턴트 구성 요소

Agent Bricks로 구축된 지식 어시스턴트(Knowledge Assistant)는 내부를 알 수 없는 '블랙박스'가 아니라, Databricks 고유의 아키텍처로 구성된 시스템입니다. 디버깅과 거버넌스를 위해서는 이러한 구성 요소를 이해하는 것이 매우 중요합니다.

<!-- <img src="../Includes/images/05-agent-bricks-components.png"/> -->

![05-agent-bricks-components](https://files.training.databricks.com/binder/prod_main/building-retrieval-agents-on-databricks-ko_kr-1.0.1/images/05-agent-bricks-components.png)

*그림 2: 에이전트 브릭스 지식 어시스턴트 구성 요소. 이 구성 요소들은 백그라운드에서 작동하며 사용자가 직접 관리할 필요가 없습니다.*


### D1. 데이터 수집 및 파싱(Data Ingestion and Parsing)

지식 어시스턴트의 기반은 **Unity Catalog 볼륨**에 저장된 데이터입니다.

* **소스(Source):** 사용자가 파일(PDF, DOCX, HTML)이 포함된 볼륨을 선택합니다. 
* **파싱(Parsing):** 시스템은 복잡한 문서에서 텍스트, 표, 이미지를 추출하기 위해 **ai\_parse\_document**, AI 함수를 사용합니다. 이를 통해 PDF의 시각적 요소(예: 차트)가 LLM이 이해할 수 있는 컨텍스트로 변환됩니다.

### D2. Databricks AI Search

파싱이 완료되면 데이터를 검색하기 위해 인덱싱되어야 합니다.

* **관리형 임베딩(Managed Embeddings):** 에이전트 브릭스는 임베딩 모델(예: GTE)을 자동으로 선택하고 **Databricks AI Search** 인덱스를 프로비저닝합니다.  
* **동기화(Synchronization):** 인덱스는 완전 관리형으로 운영됩니다. 소스 볼륨에 새로운 파일이 추가되면 벡터 검색 인덱스가 자동으로 업데이트되므로, 사용자가 수동으로 다시 인덱싱하지 않아도 에이전트가 항상 최신 지식을 보유할 수 있습니다.

### D3. 추론 엔진 및 모델 서빙(The Reasoning Engine and Model Serving)

에이전트 로직은 **Model Serving**에서 호스팅됩니다.

* **추론(Inference):** 사용자가 질문을 하면 시스템은 질문을 벡터로 변환하고, 벡터 검색에서 관련 청크(chunk)를 가져와 LLM에 전달합니다.
* **인용구 표시(Citation):** 지식 어시스턴트는 인용구를 제공하도록 설계되었습니다. 답변을 Unity Catalog 볼륨의 구체적인 소스 파일과 매핑하여 사용자가 정확성을 검증할 수 있도록 지원합니다.

### D4. 품질 루프 - 리뷰 앱 및 평가(The Quality Loop: Review App & Evaluation)

이 부분은 Agent Bricks만의 차별화된 장점입니다.

* **리뷰 앱(Review App):** 이해관계자(도메인 전문가)가 에이전트와 대화하고 피드백(좋아요/싫어요/수정)을 제공할 수 있는 내장 UI입니다.
* **LLM 평가 모델(LLM Judges):** 스템은 Mosaic AI **에이전트 평가(Agent Evaluation)** 를 사용하여 상호작용 추적 내역을 대상으로 'LLM 평가 모델'을 실행합니다. 이 평가 모델은 '충실도'(모델이 환각 현상을 일으켰는가?) 및 '정확성'과 같은 지표를 평가합니다.
* **최적화(Optimization):** Agent Bricks는 수집된 피드백을 활용하여 시스템 지침이나 구성을 업데이트하도록 제안함으로써 성능 지표를 개선합니다.




## E. 요약

**Agent Bricks를 활용한 지식 어시스턴트**는 AI 구성 요소를 수동으로 설계하는 방식에서 AI 결과물을 관리하는 방식으로의 전환을 의미합니다. 선언적 접근 방식을 활용함으로써 팀은 기업 데이터에 기반한 RAG(검색 강화 생성) 시스템을 단 몇 분 만에 배포할 수 있습니다.

**핵심 내용:**

1. **구성보다 최적화:** Agent Bricks는 모델과 검색 매개변수 선택을 자동화하여 비용과 품질 간의 균형을 맞춥니다.  
2. **통합 아키텍처:** Unity catalog 볼륨, ai\_parse\_document, 및 AI Search를 자동으로 조정합니다.
3. **피드백 기반:** 시스템은 Review App과 Agent Learning from Human Feedback(ALHF)을 통해 시간이 지남에 따라 지속적으로 개선되며, 전문가의 피드백을 시스템 개선으로 전환합니다.
