# Databricks와 MLFlow로 AI 에이전트 시스템 구축 및 평가

<img src="https://github.com/databricks-demos/dbdemos-resources/blob/main/images/product/ai-agent/agent-demo-0.png?raw=true" width="800px" style="float: right">

Databricks를 활용하면 데이터를 손쉽게 조작할 수 있는 사용자 정의 에이전트와 툴을 만들고, 외부 시스템(OpenAI SDK, Claude 에이전트, MCP 서버 등)과 연결할 수 있습니다.

하지만 에이전트 구축은 비즈니스 성과를 내기 위한 첫 걸음에 불과합니다!

이번 데모에서는 고객 지원팀을 위한 에이전트를 만들어 볼 것입니다. 고객 문의를 처리하고, 안내를 제공하여 **해결까지의 시간을 단축하고, 만족도를 높이며 비용을 절감**할 수 있습니다.

## 3/ 지식 기반 추가

에이전트가 잘 동작하지만 내부 제품 및 절차와 관련된 질문에는 답변하지 못합니다.

PDF에서 정보를 추출하고 벡터 검색 기능을 추가해, 프롬프트에 맞춤형 정보를 더해줍시다!

물론 평가를 반복하여 꾸준히 개선하는 것도 잊지 마세요.

[03-knowledge-base-rag/03.1-pdf-rag-tool]($./03-knowledge-base-rag/03.1-pdf-rag-tool) 노트북을 열어 지식 기반을 추가하세요.
